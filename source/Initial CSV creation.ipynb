## Parsing Column Meta-data from Dataset Metadata File

```python
col_name = data.columns
description = []
dataTypeName = []
for column in meta_data['columns']:
    description.append(column['description'])
    dataTypeName.append(column['dataTypeName'])

temp_d = {
            'columnName': list(col_name), 
            'description': description, 
            'dataTypeName': dataTypeName
        }

metadata_df = pd.DataFrame(temp_d)
del(temp_d, description, dataTypeName, col_name)
metadata_df
```

```python
metadata_df['dataset_id'] = 1
metadata_df
```

```python
col_unique_counts = data.nunique()
col_unique_counts
```

```python
col_unique_counts_df = col_unique_counts.to_frame(name="unique_values_count")
# converting index to column
col_unique_counts_df.reset_index(level=0, inplace=True)
col_unique_counts_df
```

```python
metadata_df2 = pd.merge(metadata_df, col_unique_counts_df, left_on = "columnName", \
                             right_on = "index", how = "left")
metadata_df2
```

```python
col_na_counts = data.isna().sum()
col_na_counts
```

```python
col_na_counts_df = col_na_counts.to_frame(name="na_values_count")
# converting index to column
col_na_counts_df.reset_index(level=0, inplace=True)
col_na_counts_df
```

```python
metadata_df3 = pd.merge(metadata_df2, col_na_counts_df, left_on = "columnName", \
                             right_on = "index", how = "left")
metadata_df3
```

```python
# dropping unnecessary columns
metadata_df3.drop(['index_x', 'index_y'], axis = 1, inplace=True)
metadata_df3.head()
```

```python
# write to processed file
metadata_df3.to_csv('./processed_csv/dat_variable_details.csv')
```

# Creating Tags CSV File

```python
# Tags table
tags_list = meta_data["tags"]
tags_df = pd.DataFrame(columns=["tag_id", "tag"])
for tag_id, tag in enumerate(tags_list):
    d = {
            "tag_id": tag_id, 
            "tag": tag
        }
    tags_df = tags_df.append(d, ignore_index = True)
tags_df
```

```python
tags_df.to_csv('./processed_csv/tags.csv')
```

```python
# Tag map
tag_map = pd.DataFrame(columns=["dataset_id", "tag_id"])
for tag_id, tag in enumerate(tags_list):
    d = {
            "tag_id": tag_id,
            "dataset_id": 1
        }
    tag_map = tag_map.append(d, ignore_index = True)
tag_map
```

```python
tag_map.to_csv('./processed_csv/tag_map.csv')
```
